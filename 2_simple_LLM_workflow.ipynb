```python
from langgraph.graph import StateGraph,START,END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import  load_dotenv
```

```python
load_dotenv()
```

```python
chat_model=ChatOpenAI()
```

```python
#create LLm State
class LLMState(TypedDict):
    question:str
    answer:str
```

```python
def model(state:LLMState)->LLMState:
    question=state["question"]
    prompt=f"Answer the following question {question}"
    answer=chat_model.invoke(prompt).content
    state['answer']=answer
    return state
```

```python
#create our graph
workflow=StateGraph(LLMState)

#add node
workflow.add_node('model',model)
#add edges
workflow.add_edge(START,'model')
workflow.add_edge('model',END)

flow=workflow.compile()
```

```python
initial_state={'question':'how stock market matured over the period'}
final_state=flow.invoke(initial_state)
print(final_state['answer'])
```
