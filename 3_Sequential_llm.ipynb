```python
from langgraph.graph import StateGraph,START,END
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict

load_dotenv()
```

```python
model=ChatOpenAI()
```

```python
class SeqFlow(TypedDict):
    title:str
    outline:str
    summary:str
```

```python
def outline_generate(state:SeqFlow)->SeqFlow:
    title=state["title"]
    prompt=f"create a detailed outline for {title}"
    outline=model.invoke(prompt).content
    state["outline"]=outline
    return state
```

```python
def summary_generate(state:SeqFlow)->SeqFlow:
    outline=state["outline"]
    prompt=f"create a summary of the detailed {outline}"
    summary=model.invoke(prompt).content
    state["summary"]=summary
    return state
```

```python
my_graph=StateGraph(SeqFlow)
#adding the nodes
my_graph.add_node('outline',outline_generate)
my_graph.add_node('summary',summary_generate)
#creating edges
my_graph.add_edge(START,'outline')
my_graph.add_edge('outline','summary')
my_graph.add_edge('summary',END)
#compiling the graph
workflow=my_graph.compile()

```

```python
initial_state={'title':"AI in India"}
final_state=workflow.invoke(initial_state)
print(final_state["outline"])
```
